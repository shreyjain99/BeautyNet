```python
import numpy as np
import pandas as pd
```

```python
data_path_scut = r"C:\Users\shrey\Downloads\SCUT-FBP5500_v2.1\SCUT-FBP5500_v2\train_test_files\All_labels.txt"
data_path_mebeauty = r"C:\Users\shrey\Downloads\mebeautydataset\generic_scores_all.xlsx"
data_path_hotornot = r"C:\Users\shrey\Downloads\hotornotdataset\eccv2010_split1.csv"
```

```python
df_scut = pd.read_csv(data_path_scut, delim_whitespace=True, header=None, names=['filename', 'score'])
```

```python
df_mebeauty = pd.read_excel(data_path_mebeauty)
```

```python
df_mebeauty = df_mebeauty[['image','mean']]
```

```python
df_mebeauty
```

```python
df_hrn = pd.read_csv(data_path_hotornot, header=None, names=['filename', 'score','tr'])
```

```python
df_hrn = df_hrn[['filename', 'score']]
```

```python
df_mebeauty['mean_5'] = 1 + (df_mebeauty['mean'] - 1) * (4 / 9)
```

```python
df_hrn['score_5'] = 1 + ((df_hrn['score'] + 3.29) / 6.81) * 4
```

```python
df_mebeauty.drop('mean',axis=1,inplace= True)
df_hrn.drop('score',axis=1,inplace= True)
```

```python
df_mebeauty.rename(columns={"image": "filename", "mean_5": "score"},inplace=True)
df_hrn.rename(columns={ "score_5": "score"},inplace=True)
```

```python
frames = [df_scut, df_mebeauty, df_hrn]

df_main = pd.concat(frames, axis = 0, ignore_index=True)
```
