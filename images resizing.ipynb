```python
## SCUT IMAGE RESIZED

from PIL import Image

input_folder = img_path_scut
output_folder = r"C:\Users\shrey\Downloads\images_combined"
desired_size = (224, 224)  # Replace width and height with the desired dimensions

# Create the output folder if it doesn't exist
os.makedirs(output_folder, exist_ok=True)

# Loop through all the files in the input folder
for filename in os.listdir(input_folder):
    # Check if the file is an image (you can add more extensions if needed)
    if filename.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.gif')):
        # Open the image file
        img_path = os.path.join(input_folder, filename)
        with Image.open(img_path) as img:
            # Resize the image
            img_resized = img.resize(desired_size, Image.Resampling.LANCZOS)
            # Save the resized image to the output folder
            output_path = os.path.join(output_folder, filename)
            img_resized.save(output_path)

print("All images have been resized and saved to the output folder.")
```

```python
import os
import pandas as pd

# Define paths
folder_path = r"C:\Users\shrey\Downloads\images_combined"
csv_path = r"C:\Users\shrey\Downloads\data_combined_new.csv"

# Load the CSV file
df = pd.read_csv(csv_path)

# Extract the list of image names from the CSV (assuming the image names are in a column named 'image_name')
csv_image_names = df['filename'].tolist()

# List all image files in the folder
folder_image_names = [f for f in os.listdir(folder_path) if f.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.gif'))]

# Remove image from folder if name not in CSV
for image_name in folder_image_names:
    if image_name not in csv_image_names:
        os.remove(os.path.join(folder_path, image_name))
        print(f'Removed {image_name} from folder.')

# Remove name from CSV and image from folder if name exists in CSV but image does not exist in folder
csv_image_names_to_remove = []
for image_name in csv_image_names:
    if image_name not in folder_image_names:
        csv_image_names_to_remove.append(image_name)

# Update the DataFrame to remove the entries not existing in the folder
df = df[~df['image_name'].isin(csv_image_names_to_remove)]


print(f'Removed entries from CSV: {csv_image_names_to_remove}')
```
